```python
def add_():
    return '''
def add(a, b):
    return a+b
'''

def fancy_func_():
    return '''
def fancy_func(a, b, c, d):
    e = add(a, b)
    f = add(c, d)
    g = add(e, f)
    return g
'''

def evoke_():
    return add_() + fancy_func_() + 'print(fancy_func(1, 2, 3, 4))'
```

```python
prog = evoke_()
print(prog)
y = compile(prog, '', 'exec')
exec(y)
```

```python
import torch
from torch import nn
from d2l import torch as d2l

def get_net():
    net = nn.Sequential(
        nn.Linear(512, 256),
        nn.ReLU(),
        nn.Linear(256, 128),
        nn.ReLU(),
        nn.Linear(128, 2)
    )
    return net

x = torch.randn(size=(1, 512))
net = get_net()
net(x)
```

```python
net = torch.jit.script(net)
net(x)
```
