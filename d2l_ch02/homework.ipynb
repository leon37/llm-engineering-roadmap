```python
#题目：手动实现一个“单层线性网络”的前向传播和反向传播
# 需求描述：
# 创建一个输入张量 X，形状是 (2, 3)（代表 batch_size=2，特征数=3），值随机。
# 创建一个权重张量 w，形状是 (3, 1)，值随机，且需要计算梯度。
# 创建一个偏置标量 b，初始值为 0，且需要计算梯度。
# 计算逻辑：实现公式
# Y
# =
# X
# w
# +
# b
# Y=Xw+b
# 。
# Loss计算：计算
# Y
# Y
#  所有元素的均值作为 Loss。
# 求导：对 Loss 进行反向传播。
# 验证：打印 w 的梯度形状，确认它是否等于 w 的形状。
```

```python
import torch
X = torch.randn((2, 3))
X
```

```python
w = torch.randn((3,1), requires_grad=True)
b = torch.zeros(1, requires_grad=True)
```

```python
Y = torch.mm(X, w) + b
Y.mean().backward()
w.grad.shape== w.shape
w.grad.shape == w.shape
```
